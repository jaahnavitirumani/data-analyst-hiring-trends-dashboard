```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("andrewmvd/data-analyst-jobs")

print("Path to dataset files:", path)
```

```python
#install libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import os

file_path = os.path.join(path, "DataAnalyst.csv")

df = pd.read_csv(file_path)

df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
df.head()
```

```python
#missing values
df.isnull().sum()
```

```python
#find top hiring companies
df['Company Name'].value_counts().head(10)
```

```python
#visualization
import matplotlib.pyplot as plt

df['Company Name'].value_counts().head(10).plot(kind='bar')

plt.title('Top 10 Hiring Companies')
plt.show()
```

```python
df['Salary Estimate'].head()
```

```python
#job location Analysis
df['Location'].value_counts().head(10)
```

```python
#chart
df['Location'].value_counts().head(10).plot(kind='bar')
plt.title('Top Hiring Locations')
plt.show()
```

```python
df[df['Location'].str.contains('Remote', case=False, na=False)].shape[0]
```

```python
skills = ['SQL', 'Python', 'Excel', 'Tableau', 'Power BI', 'R']

for skill in skills:
    count = df['Job Description'].str.contains(
        skill,
        case=False,
        na=False
    ).sum()

    print(skill, count)
```

```python
df.to_csv("cleaned_jobs.csv", index=False)
```
